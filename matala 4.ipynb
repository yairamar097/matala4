```python
#יבוא והתקנת סיפריות מבוקשות
!pip install beautifulsoup4
!pip install openpyxl
import requests
from bs4 import BeautifulSoup
import urllib.parse
import re
import pandas as pd
```

```python
#פונקציה שמפצלת כתובת לרחוב ומספר דירה ברחוב 
def split_words_and_numbers(text):
    # Use regular expressions to split the text into words and numbers
    words = re.findall(r'\b[a-zA-Zא-ת]+\b', text)
    numbers = re.findall(r'\b\d+\b', text)
    
    # Join the words and numbers into a single string with spaces between them
    street = ' '.join(words)
    number_in_street = ' '.join(numbers)
    
    return street , number_in_street
split_words_and_numbers(' 345  יאיר עמר')
```

```python
#פונקציה שמקבלת עיר ומחזירה רשימת מילונים שמורכבים מפרטים לגבי כל מודעה ורשימת קישורים  
def add(city):
    add_list = list()
    lst = []
    add_url = []
    import requests
    from bs4 import BeautifulSoup
     
    url = 'https://www.madlan.co.il/for-sale/' + urllib.parse.quote(city)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}  # This is chrome, you can set whatever browser you like
    html = requests.get(url,headers= headers)
    soup = BeautifulSoup(html.content,'html.parser')
    try:
        
        
        adds = soup.select('[data-auto=listed-bulletin]')
        for add in adds:
            try:
                area = add.find_all('div', class_="css-1i00gl6 e1sx3tzs17")[1].get_text()
            except:
                area = ''
                
            try:
                type_ = add.find_all('div', class_="css-1i00gl6 e1sx3tzs17")[2].get_text()
            except:
                type = ''
                
            try:
                x = add.find_all('div', class_="css-1i00gl6 e1sx3tzs17")[3].get_text()
            except:
                x = ''
            
            try:
                x = x.split(",")
                city_area = x[1]
                street , number_in_street = split_words_and_numbers(x[0])
            except:
                x = ''
                
            try:
                price = add.find('div', class_="css-19tpcfu eu6esji3").get_text()
            except:
                price = ''
          
            try:
                link = add.find('a')['href']
                add_link = 'https://www.madlan.co.il' + link
            except:
                link = ''
                add_link = ''  
                    
           
            add_url.append([add_link])
            dict={}
            dict['type']=type_
            dict['area'] = area
            dict['street'] = street
            dict['city_area'] = city_area
            dict['number_in_street'] = number_in_street
            dict['price'] = price
            dict['add_link'] = add_link
            
            lst.append(dict)
            
        return  lst ,add_url
    except:
        return None ,'bbb'
```
